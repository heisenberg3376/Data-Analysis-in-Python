# Pandas series

```python
import numpy as np

ser = pd.Series([1,5,6,4,0])
ser1 = np.array([1,5,6,4,0])
print(ser, ser1)
# pandas Series are internally numpy arrays
```

# Which Borough has most Illegal Parking Complaints

```python
illPark = illPark_comp['Borough'].value_counts()
print(illPark) # this is a Pandas Series
illPark.plot(kind='bar')

# it's 'Brooklyn'
```

```python
comp_count = comp['Borough'].value_counts()
comp_ratio = illPark / comp_count
comp_ratio.plot(kind='bar')
```

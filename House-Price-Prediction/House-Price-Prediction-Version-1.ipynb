# Reading the Train and Test Data

```python
import pandas as pd

# used to display all the rows and columns of a dataframe
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)

# reading the train,test data
train = pd.read_csv(r'C:\Users\Katta\OneDrive\Desktop\DataSets\train.csv')
test = pd.read_csv(r'C:\Users\Katta\OneDrive\Desktop\DataSets\test.csv')

# we store the test id's for further use
test_ids = test['Id']

# dropping the columns with more (>1000) NaN values
test.drop(['Alley','PoolQC','Fence','MiscFeature'],axis=1,inplace=True)
train.drop(['Alley','PoolQC','Fence','MiscFeature'],axis=1,inplace=True)
```

```python
train.head()
```

# Filling the NaN values with most Frequent values of that column

```python
for col in train.columns:
    train[col].fillna(train[col].mode()[0],inplace=True)
    
for col in test.columns:
    test[col].fillna(test[col].mode()[0],inplace=True)
```

```python
#train.info()  
```

# Encoding the Categorical Data into Numerical Data

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cols = ['MSZoning','Street','LotShape','LandContour','Utilities','LotConfig','LandSlope','Neighborhood','Condition1','Condition2','BldgType','HouseStyle','RoofStyle',
       'RoofMatl','Exterior1st','Exterior2nd','MasVnrType','ExterQual','ExterCond','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Heating',
       'HeatingQC','CentralAir','Electrical','KitchenQual','Functional','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PavedDrive','SaleType','SaleCondition']
def encode(df):
    df.drop('Id',axis=1,inplace=True)
    for i in cols:
        df[i] = le.fit_transform(df[i])

    return df

train = encode(train)
test = encode(test)
```

# Importing the Random Forest Regressor 

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

# Defining the Features, Labels, and classifier

```python
X = train.drop('SalePrice',axis=1)
y = train.SalePrice

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
clf = RandomForestRegressor(random_state=29)
# training the classifier
clf.fit(X_train,y_train)
# checking it's accuracy
clf.score(X_test,y_test)
```

Predicting the SalePrice for Given Test DataFrame

```python
preds = clf.predict(test)

sub_df = pd.DataFrame({'Id':test_ids,'SalePrice':preds})
sub_df.to_csv(r'C:\Users\Katta\OneDrive\Desktop\DataSets\sub_df.csv',index=False)
sub_df
```

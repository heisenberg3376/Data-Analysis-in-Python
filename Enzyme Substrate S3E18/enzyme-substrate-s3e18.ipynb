```python
from sklearn.preprocessing import StandardScaler
def clean(df):
    sc = StandardScaler()
    df.iloc[:,:32] = sc.fit_transform(df.iloc[:,:32])
    df = df.drop('id',axis=1)
    return df
```

```python
train = clean(train)
```

```python
test = clean(test)
```

```python
X = train.iloc[:,:31]
y1 = train.iloc[:,31]
y2 = train.iloc[:,32]
```

<div style="color:white;
           display:fill;
           border-radius:5px;
           background-color:#4605fa;
           font-size:150%;
           font-family:Verdana;
           letter-spacing:0.5px;">
    <b style="text-align:center;padding-left:38%">Splitting the Data</b>
</div>

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y1_train,y1_test = train_test_split(X,y1,test_size=0.2,random_state=33)
X_train,X_test,y2_train,y2_test = train_test_split(X,y2,test_size=0.2,random_state=33)
```

<div style="color:white;
           display:fill;
           border-radius:5px;
           background-color:#4605fa;
           font-size:150%;
           font-family:Verdana;
           letter-spacing:0.5px;">
    <b style="text-align:center;padding-left:18%">Creating two XGBoost Models for 2 target columns</b>
</div>

```python
import xgboost as xgb

xgb_model_1 = xgb.XGBClassifier(n_estimators=50,subsample=0.8, learning_rate=1.0, max_depth=2, random_state=42)
xgb_model_2 = xgb.XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
```

<div style="color:white;
           display:fill;
           border-radius:5px;
           background-color:#4605fa;
           font-size:150%;
           font-family:Verdana;
           letter-spacing:0.5px;">
    <b style="text-align:center;padding-left:38%">Training the Models</b>
</div>

```python
xgb_model_1.fit(X_train,y1_train)
```

```python
xgb_model_2.fit(X_train,y2_train)
```

```python
print(xgb_model_1.score(X_test,y1_test))
print(xgb_model_2.score(X_test,y2_test))
```

```python
preds1 = xgb_model_1.predict(test)
preds2 = xgb_model_2.predict(test)
```
